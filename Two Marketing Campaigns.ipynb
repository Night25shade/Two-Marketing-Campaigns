```python
# Convert the 'Week' column to datetime format for time series analysis
df['Week'] = pd.to_datetime(df['Week'])
```

```python
# Remove dollar signs and commas from the 'Payment' column and convert to numeric
df['Payment'] = df['Payment'].replace('[\$,]', '', regex=True).astype(float)

# Calculate Returns Percentage
df['Returns %'] = (df['Payment'] / df['Cost']) * 100
```

```python
# Calculate weekly metrics
df_weekly = df.resample('W', on='Week').sum()

# Calculate monthly metrics
df_monthly = df.resample('M', on='Week').sum()

# Calculate Returns Percentage for both weekly and monthly data
df_weekly['Returns %'] = (df_weekly['Payment'] / df_weekly['Cost']) * 100
df_monthly['Returns %'] = (df_monthly['Payment'] / df_monthly['Cost']) * 100
```

```python
# Calculate weekly averages for impressions, clicks, and prospects
df_weekly['Average Impressions'] = df_weekly['Impressions'] / df_weekly['Impressions'].count()
df_weekly['Average Clicks'] = df_weekly['Clicks'] / df_weekly['Clicks'].count()
df_weekly['Average Prospects'] = df_weekly['Prospects'] / df_weekly['Prospects'].count()

# Export updated weekly data to Excel
df_weekly.to_excel('weekly_metrics_updated.xlsx')
```

```python
# Calculate returns percentage
df['Returns %'] = (df['Payment'] / df['Cost']) * 100
```

```python
# Group the data by 'Campaign' and calculate average metrics for each channel
Campaign_avg_metrics = df.groupby('Campaign').agg({
    'Cost': 'mean',
    'Impressions': 'mean',
    'Clicks': 'mean',
    'Prospects': 'mean',
    'Payment': 'mean',
    'Returns %': 'mean'
}).reset_index()

# Group the data by 'Search_Keyword' and calculate average metrics for each Search_Keyword
Search_Keyword_avg_metrics = df.groupby('Search_Keyword').agg({
    'Cost': 'mean',
    'Impressions': 'mean',
    'Clicks': 'mean',
    'Prospects': 'mean',
    'Payment': 'mean',
    'Returns %': 'mean'
}).reset_index()
```
